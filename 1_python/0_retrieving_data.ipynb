```python
# IMPORT THE REQUIRED LIBRARIES
import kagglehub
import os
import pandas as pd

# DOWNLOAD THE DATASET FROM KAGGLE
path = kagglehub.dataset_download("mashlyn/online-retail-ii-uci")

files = os.listdir(path)
file_path = os.path.join(path, files[0])

df = pd.read_csv(file_path)

print(df.head())

# CREATE DATA FOLDER IN THIS REPOSITORY AND SAVE THE DATASET
os.makedirs("../0_data/0_raw", exist_ok=True)

df.to_csv("../0_data/0_raw/online_retail_II.csv", index=False)
```
